```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from typing import List, Dict, Tuple, Union
from numpy.typing import ArrayLike
import matplotlib.pyplot as plt
from nn import *
```

Here I am going to pull in the digits dataset

```python
digits = load_digits()
print(digits.data.shape)
```

Here I split into train, test, and validation set

```python
# Split off Final test set
X_train, X_test, y_train, y_test = train_test_split(digits["data"], digits["data"], test_size=0.2, random_state=13)

# Split intial train into train and validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, X_train, test_size=0.2, random_state=13)

X_val
```

Now I want to spin up a Nueral net and train the data

```python
auto_encoder=NeuralNetwork([{'input_dim': 64, 'output_dim': 16, 'activation': 'relu'}, {'input_dim': 16, 'output_dim': 64, 'activation': 'relu'}],lr=.001,seed=13,batch_size=10,epochs=100,loss_function="mean_squared_error")
per_epoch_loss_train,per_epoch_loss_val=auto_encoder.fit(X_train,y_train,X_val,y_val)
```

```python
epochs = range(1, len(per_epoch_loss_train) + 1)
# Step 1: Predict Reconstruction
reconstructed_output = auto_encoder.predict(X_val)

# Step 2: Calculate Error
reconstruction_error = np.mean(np.square(X_val - reconstructed_output))

# Step 3: Print or Use the Average Error
print("Average Reconstruction Error:", reconstruction_error)
plt.plot(epochs, per_epoch_loss_train, 'b', label='Training loss')
plt.plot(epochs, per_epoch_loss_val, 'r', label='Validation loss')
plt.title('Training and Validation loss per epoch')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
